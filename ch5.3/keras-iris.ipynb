```python
import tensorflow as tf 
import tensorflow.contrib.keras as keras
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


iris_data = pd.read_csv("iris.csv", encoding="utf-8")
y_labels = iris_data.loc[:,"Name"]
x_data = iris_data.loc[:,["SepalLength","SepalWidth","PetalLength","PetalWidth"]]

# tensorflowでの表現のためone-hotベクトル化
labels = {
    'Iris-setosa': [1,0,0],
    'Iris-versicolor':[0,1,0],
    'Iris-virginica':[0,0,1]
}
# nameラベルのリストが、ベクトルの配列に
y_nums = np.array(list(map(lambda v: labels[v], y_labels)))
x_data = np.array(x_data)

# 学習用とテスト用にデータ分割
x_train, x_test, y_train, y_test = train_test_split(x_data, y_nums,train_size = 0.8)

# モデル構造定義
Dense = keras.layers.Dense
# 順番にニュートラルネットワークに追加するモデル作成
model = keras.models.Sequential()
# 　Dense(10,活性化関数,input_shape=入力次元数)：ユニットが１０あるネットワーク作成
# relu 入力した値が0以下のとき0になり、1より大きいとき入力をそのまま出力
model.add(Dense(10, activation='relu', input_shape=(4,)))
# 出力次元数３でモデル追加
# softmaxは出力層で分類に用いられる
model.add(Dense(3, activation='softmax'))

# モデル構築
#  最適化adam法　Adaptive Moment Estimationそれぞれのパラメータに対し学習率を計算し適応
# 損失関数（誤差の評価)　カテゴリ分類の評価にはcategorical_crossentropy
# mterics 評価関数のリスト
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])

# 学習を実行
# batch_size：一回に計算するデータの数、epochs：繰り返し回数
model.fit(x_train, y_train, batch_size=20, epochs=300)

# モデル評価 verbose=1(0-2)：標準ログ出力
score = model.evaluate(x_test, y_test, verbose = 1)
print("正解率",score[1], 'loss=',score[0])
```
